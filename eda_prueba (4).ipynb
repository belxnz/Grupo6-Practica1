```python
# Matriz de correlación:
import seaborn as sns
import matplotlib.pyplot as plt

corrmat = df.corr()
plt.figure(figsize=(20,20))
sns.heatmap(corrmat, vmax=.8, square=True)

# Identificamos las variables que vamos a quitar (>0.95)
to_drop = set()
for i in range(len(corrmat.columns)):
    for j in range(i):
        if abs(corrmat.iloc[i, j]) > 0.95:
            colname = corrmat.columns[i]
            to_drop.add(colname)

# Hacemos drop sobre esas variables
df_train = df.drop(to_drop, axis=1)

# Imprimimos el segundo heatmap
corrmat = df_train.corr()
plt.figure(figsize=(20,20))
sns.heatmap(corrmat, vmax=.8, square=True)
```

En el primer heat map se puede observar que existen varias variables con correlación muy alta, tanto positiva como negativa. Para evitar la presencia de datos dependientes, es conveniente eliminar aquellas variables que presentan una correlación elevada con otras. Para ello, se debe prestar atención a los cuadrados del heatmap que sean muy oscuros o muy blancos. Es importante tener en cuenta que una variable siempre tendrá una alta correlación consigo misma, por lo que debemos buscar aquellas variables distintas con alta correlación.
Hemos decidido eliminar una de las variables de los pares de variables cuya correlación es mayor que 0.95, tanto correlación positiva como negativa y por tanto nos hemos quedado con solo 25 variables restantes.

```python
corr = df_train.corr()
corr[['salida']].sort_values(by = 'salida',ascending = False).style.background_gradient()
```

En las matrices de correlación se puede observar lo siguiente:
1. Las variables dswrf_sfc son las que más correlacion tiene con la variable 'salida' y se corresponden a la media de radiación de onda corta que llega del cielo en la superficie, más específicamente a las horas 2,3,4,5 (12h, 15h, 18h y 24h respectivamente)
2. Las variables uswrf_sfc que corresponde a la radiación de onda corta que procede del suelo sobre la superficie a las horas 2 y 3 también tienen un alto nivel de correlación.
3. A la hora 5, la variable ulwrf_sfc que corresponde a la radiación de onda larga que procede del suelo sobre la superficie tiene alta correlación con la salida.
4. Podemos destacar como la variables más correlacionada positiva es mucho mayor que la negativa (|0.87|>|-0.37|)
5. No se observa que exista ninguna variable adicional con una correlación demasiado alta, por lo tanto no es necesario eliminar ninguna más.

```python
# Dividir datos en train y test
df_train = df[0:438]
df_test = df[438:]
```

Hemos dividido los datos en df_train y df_test, el primero contiene 10 años de datos mientras que el segundo los 2 años restantes. Vamos a trabajar con df_train por ahora y más tarde a la hora de evaluar, df_test.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split

# Dividir el conjunto de datos en conjunto de entrenamiento y conjunto de prueba
x_train, x_test, y_train, y_test = train_test_split(df_train.drop('salida', axis=1), df_train['salida'], test_size=0.2)

# Instanciar el clasificador KNN con hiperparámetros por omisión
knn = KNeighborsRegressor()

# Entrenar el clasificador en el conjunto de entrenamiento
knn.fit(x_train, y_train)

# Evaluar el rendimiento del clasificador en el conjunto de prueba
accuracy = knn.score(x_test, y_test)
print(f'Precisión del clasificador KNN: {accuracy:.2f}')
```

```python
from sklearn import neighbors
from sklearn.metrics import mean_squared_error 
from math import sqrt
import matplotlib.pyplot as plt
%matplotlib inline

rmse_val = [] #to store rmse values for different k
for K in range(20):
    K = K+1
    model = neighbors.KNeighborsRegressor(n_neighbors = K)

    model.fit(x_train, y_train)  #fit the model
    pred=model.predict(x_test) #make prediction on test set
    error = sqrt(mean_squared_error(y_test,pred)) #calculate rmse
    rmse_val.append(error) #store rmse values
    print('RMSE value for k= ' , K , 'is:', error)

#plotting the rmse values against k values
curve = pd.DataFrame(rmse_val) #elbow curve 
curve.plot()
```

```python
from sklearn import tree
# Here, we define the type of training method (nothing happens yet)
clf = tree.DecisionTreeClassifier()
# Now, we train (fit) the method on the (X,y) dataset
clf.fit(x_train, y_train)
# clf **has been changed** and now contains the trained model
text_representation = tree.export_text(clf)
print(text_representation)
```
